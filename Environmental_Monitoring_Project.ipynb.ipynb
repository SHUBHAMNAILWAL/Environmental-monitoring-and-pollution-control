```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (8,4)

df = pd.read_csv("C:\\Users\\nailw\\Downloads\\archive (1).zip")
print("Dataset shape:", df.shape)
df.head()
```

```python
df.info()
```

```python
df['Datetime'] = pd.to_datetime(df['Date'].astype(str) + ' ' + df['Time'].astype(str), errors='coerce')
df = df.drop(columns=['Date','Time'])
df = df.loc[df['Datetime'].notna()].reset_index(drop=True)

pollutants = ['CO(GT)','NOx(GT)','NO2(GT)','O3(GT)','SO2(GT)','PM2.5','PM10']
for p in pollutants:
    df.loc[df[p] < 0, p] = np.nan
df['Year'] = df['Datetime'].dt.year
df['Month'] = df['Datetime'].dt.month
df['DateOnly'] = df['Datetime'].dt.date
df.head()
```

```python
df[pollutants].describe().T
```

```python
daily = df.groupby('DateOnly')[['PM2.5','PM10']].mean().reset_index()
plt.figure()
plt.plot(daily['DateOnly'], daily['PM2.5'], label='PM2.5')
plt.plot(daily['DateOnly'], daily['PM10'], label='PM10')
plt.xlabel('Date')
plt.ylabel('Concentration')
plt.title('Daily Mean PM2.5 and PM10')
plt.legend()
plt.tight_layout()
plt.savefig("C:\\Users\\nailw\\Downloads\\daily_pm_trend.png")
plt.close()
print('Saved plot: daily_pm_trend.png')
```

```python
plt.figure()
df[['PM2.5','PM10','NO2(GT)','O3(GT)','SO2(GT)']].boxplot()
plt.title('Pollutant Distributions')
plt.ylabel('Concentration')
plt.tight_layout()
plt.savefig("C:\\Users\\nailw\\Downloads\\pollutant_boxplots.png")
plt.close()
print('Saved plot: pollutant_boxplots.png')
```

```python
monthly = df.groupby('Month')['PM2.5'].mean().reindex(range(1,13))
plt.figure()
monthly.plot(marker='o')
plt.xlabel('Month')
plt.ylabel('Average PM2.5')
plt.title('Average Monthly PM2.5 (Seasonality)')
plt.tight_layout()
plt.savefig("C:\\Users\\nailw\\Downloads\\monthly_pm25.png")
plt.close()
print('Saved plot: monthly_pm25.png')
```

```python
top10 = df.groupby('DateOnly')['PM2.5'].mean().nlargest(10).reset_index()
top10
```
